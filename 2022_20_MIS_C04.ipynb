#### Modelos aleatorios de redes

```python
# Cargue de paquetes necesarios para la clase
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Grafo tipo Erdos-Renyi
n = 5
p = 0.5
G_ER = nx.erdos_renyi_graph(n,p) 
nx.draw(G_ER)
```

```python
# Grafo generado con el modelo de configuración (n=5)
degrees = [2,3,4,1,4] # Qué estaría mal acá?
print("¿La cantidad de grados en el listado es igual a la cantidad de nodos en la red? {}".format(len(degrees) == n))
G_CM = nx.configuration_model(degrees)
nx.draw(G_CM)
```

```python
# Limpieza del grafo anterior

# Eliminar "self-loops"
G_CM.remove_edges_from(nx.selfloop_edges(G_CM))
nx.draw(G_CM)

```
